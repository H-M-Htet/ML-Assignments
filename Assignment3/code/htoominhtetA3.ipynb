```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from sklearn.model_selection import KFold
import mlflow
import mlflow.sklearn
import os

class LogisticRegression:
    def __init__(self, regularization, k, n, lr=0.01, num_epoch=500, method='mini', batch_size=50, cv = KFold(n_splits=3), max_iter=1000, init_method='xavier', lambda_=0.1, momentum=0.0):
        self.k = k  # number of classes
        self.n = n  # number of features
        self.regularization = regularization  # enable or disable L2 regularization
        self.lr = lr  # learning rate
        self.num_epoch = num_epoch 
        self.method = method 
        self.batch_size = batch_size 
        self.cv = cv
        self.max_iter = max_iter
        self.init_method = init_method
        self.lambda_ = lambda_  # regularization strength
        self.momentum = momentum
       

    
    def accuracy(self, y_true, y_pred):
        return np.mean(y_true == y_pred)

    def precision(self, y_true, y_pred, c):
        tp = np.sum((y_pred == c) & (y_true == c))
        fp = np.sum((y_pred == c) & (y_true != c))
        return tp / (tp + fp) if (tp + fp) > 0 else 0

    def recall(self, y_true, y_pred, c):
        tp = np.sum((y_pred == c) & (y_true == c))
        fn = np.sum((y_pred != c) & (y_true == c))
        return tp / (tp + fn) if (tp + fn) > 0 else 0

    def f1_score(self, precision, recall):
        return 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    def macro_metrics(self, y_true, y_pred):
        precisions = [self.precision(y_true, y_pred, c) for c in range(self.k)]
        recalls = [self.recall(y_true, y_pred, c) for c in range(self.k)]
        f1s = [self.f1_score(p, r) for p, r in zip(precisions, recalls)]
        return np.mean(precisions), np.mean(recalls), np.mean(f1s)

    def weighted_metrics(self, y_true, y_pred):
        class_counts = np.bincount(y_true)
        total = len(y_true)
        precisions = [self.precision(y_true, y_pred, c) for c in range(self.k)]
        recalls = [self.recall(y_true, y_pred, c) for c in range(self.k)]
        f1s = [self.f1_score(p, r) for p, r in zip(precisions, recalls)]
        weights = class_counts / total
        weighted_precision = np.sum(np.array(precisions) * weights)
        weighted_recall = np.sum(np.array(recalls) * weights)
        weighted_f1 = np.sum(np.array(f1s) * weights)
        return weighted_precision, weighted_recall, weighted_f1

    def log_metrics(self, y_true, y_pred):
        acc = self.accuracy(y_true, y_pred)
        mlflow.log_metric("accuracy", acc)
        p_macro, r_macro, f1_macro = self.macro_metrics(y_true, y_pred)
        mlflow.log_metric("macro_precision", p_macro)
        mlflow.log_metric("macro_recall", r_macro)
        mlflow.log_metric("macro_f1", f1_macro)
        p_weighted, r_weighted, f1_weighted = self.weighted_metrics(y_true, y_pred)
        mlflow.log_metric("weighted_precision", p_weighted)
        mlflow.log_metric("weighted_recall", r_weighted)
        mlflow.log_metric("weighted_f1", f1_weighted)
    

    def initialize_methods(self, features): #condition checking for initializing methods
        if self.init_method == 'zero':
            return np.zeros((features, self.k))
        elif self.init_method == 'xavier':
            limit = 1.0 / np.sqrt(features)
            # lower, upper = - (1.0 / np.sqrt(features)), (1.0 / np.sqrt(features))
            return np.random.uniform(-limit, limit, size=(features, self.k))
    
    def one_hot(self, y, num_classes):
        one_hot = np.eye(num_classes)[y.astype(int).flatten()]
        if one_hot.ndim > 2:
            one_hot = one_hot.squeeze()
        return one_hot

    # print("X_val shape:", X_cross_val.shape)
    # print("y_val shape:", y_cross_val.shape)
    # print("yhat_val_probs shape:", yhat_val_probs.shape)

    def cross_entropy(self, y_true, y_pred):
        return -np.sum(y_true * np.log(y_pred + 1e-15)) / y_true.shape[0]
        

    def fit(self, X_train, Y_train):
        
        self.kfold_scores = list()
        self.val_loss_old = np.inf
        self.losses = []

        # Y_train = np.array(Y_train)  

        for fold, (train_idx, val_idx) in enumerate(self.cv.split(X_train)):
            
            X_cross_train = X_train[train_idx]
            y_cross_train = Y_train[train_idx]
            X_cross_val   = X_train[val_idx]
            y_cross_val   = Y_train[val_idx]
            
            self.theta = self.initialize_methods(X_cross_train.shape[1])
            self.prev_step = np.zeros_like(self.theta)  
            
            with mlflow.start_run(run_name=f"Fold-{fold}", nested=True):
                
                params = {"gradient": self.method, "lr": self.lr, "reg": type(self).__name__, "init_method": self.init_method, "momentum": self.momentum, "lambda_": self.lambda_}
                mlflow.log_params(params=params)
                
                for epoch in range(self.num_epoch):
                
                    perm = np.random.permutation(X_cross_train.shape[0])
                    X_cross_train = X_cross_train[perm]
                    y_cross_train = y_cross_train[perm]


                    if self.method == 'mini':

                        for batch_idx in range(0, X_cross_train.shape[0], self.batch_size):
                            X_method_train = X_cross_train[batch_idx:batch_idx+self.batch_size, :]
                            y_method_train = y_cross_train[batch_idx:batch_idx+self.batch_size]
                            train_loss = self._train(X_method_train, y_method_train)
                            self.losses.append(train_loss) 
                    else:
                        
                        X_method_train = X_cross_train
                        y_method_train = y_cross_train
                        train_loss = self._train(X_method_train, y_method_train)

                    mlflow.log_metric(key="train_loss", value=train_loss, step=epoch)

                    yhat_val_probs = self._predict(X_cross_val, self.theta)
                    val_loss_new = self.cross_entropy(self.one_hot(y_cross_val,self.k), yhat_val_probs)

                    mlflow.log_metric(key="val_loss", value=val_loss_new, step=epoch)
                    
                    if np.allclose(val_loss_new, self.val_loss_old):
                        break
                    self.val_loss_old = val_loss_new
            
                self.kfold_scores.append(val_loss_new)
                print(f"Fold {fold}: {val_loss_new}")

    def _train(self, X, y):
        m = X.shape[0]
        y_onehot = self.one_hot(y, self.k)
        yhat = self._predict(X, self.theta) #(64,4)
        train_loss = - np.sum(y_onehot * np.log(yhat)) / m

        grad = X.T @ (yhat - y_onehot) / m

        if self.regularization is not None:  
            grad += self.regularization.derivation(self.theta)

        step = self.lr * grad
        if self.momentum > 0:
            self.theta = self.theta - step + self.momentum * self.prev_step
            self.prev_step = step
        else:
            self.theta = self.theta - step

        return train_loss
            
    
    # def gradient(self, X, Y):
    #     m = X.shape[0]
    #     H = self._predict(X, self.W)
    #     loss = -np.sum(Y*np.log(H)) / m             ##cross-entropy loss
    #     grad = X.T @ (H - Y)
    #     if self.regularization:
    #         reg_term = (self.lambda_ / m) * self.W
    #         grad += reg_term
    #         loss += (self.lambda_ / (2 * m)) * np.sum(np.square(self.W))
    #     return loss, grad
    
    
    def softmax(self, h_theta):
        return np.exp(h_theta) / np.sum(np.exp(h_theta), axis=1, keepdims=True)
    
    def _predict(self, X, W):
        return self.softmax(X @ W)
    
    def predict(self, X):
        return np.argmax(self._predict(X, self.theta), axis=1)
    
    def support(self, y_true):
        classes = np.unique(y_true)
        return {c: np.sum(y_true == c) for c in classes}

    def print_classification_report(self, y_true, y_pred):
        classes = np.unique(y_true)
        print("\nCustom Classification Report:")
        print(f"{'Class':<10}{'Precision':<12}{'Recall':<10}{'F1-Score':<12}{'Support':<10}")
        for c in classes:
            p = self.precision(y_true, y_pred, c)
            r = self.recall(y_true, y_pred, c)
            f = self.f1_score(p, r)
            s = np.sum(y_true == c)
            print(f"{c:<10}{p:<12.4f}{r:<10.4f}{f:<12.4f}{s:<10}")
        acc = self.accuracy(y_true, y_pred)
        mp, mr, mf = self.macro_metrics(y_true, y_pred)
        wp, wr, wf = self.weighted_metrics(y_true, y_pred)
        print(f"\nAccuracy: {acc:.4f}")
        print(f"Macro Avg -> Precision: {mp:.4f}, Recall: {mr:.4f}, F1-Score: {mf:.4f}")
        print(f"Weighted Avg -> Precision: {wp:.4f}, Recall: {wr:.4f}, F1-Score: {wf:.4f}")

    

    # def cross_validate(self, X, Y, folds=5):
    #     kf = KFold(n_splits=folds)
    #     scores = []
    #     with mlflow.start_run(run_name="Cross-Validation"):
    #         for fold, (train_idx, test_idx) in enumerate(kf.split(X)):
    #             X_train, X_test = X[train_idx], X[test_idx]
    #             Y_train, Y_test = Y[train_idx], Y[test_idx]
    #             self.fit(X_train, Y_train)
    #             y_pred = self.predict(X_test)
    #             acc = self.accuracy(Y_test.argmax(axis=1), y_pred)
    #             scores.append(acc)
    #             mlflow.log_metric(f"accuracy_fold_{fold+1}", acc)
    #             print(f"Fold {fold+1} Accuracy: {acc}")
    #         mean_acc = np.mean(scores)
    #         mlflow.log_metric("cross_val_accuracy", mean_acc)
    #         print(f"Mean Cross-Validation Accuracy: {mean_acc}")

    # def grid_search(self, X, Y, lr_list=[0.01, 0.05, 0.1], lambda_list=[0.01, 0.1, 1]):
    #     best_score = 0
    #     best_params = {}
    #     with mlflow.start_run(run_name="Grid Search"):
    #         for lr in lr_list:
    #             for lam in lambda_list:
    #                 self.lr = lr
    #                 self.lambda_ = lam
    #                 self.fit(X, Y)
    #                 y_pred = self.predict(X)
    #                 acc = self.accuracy(Y.argmax(axis=1), y_pred)
    #                 mlflow.log_param("lr", lr)
    #                 mlflow.log_param("lambda", lam)
    #                 mlflow.log_metric("accuracy", acc)
    #                 print(f"lr: {lr}, lambda: {lam}, accuracy: {acc}")
    #                 if acc > best_score:
    #                     best_score = acc
    #                     best_params = {'lr': lr, 'lambda': lam}
    #         mlflow.log_metric("best_accuracy", best_score)
    #         mlflow.log_params(best_params)
    #         print("Best Hyperparameters:", best_params)
    #         print("Best Accuracy:", best_score)

    def plot(self):
        plt.plot(self.losses)
        plt.title("Loss Curve")
        plt.xlabel("Iterations")
        plt.ylabel("Loss")
        plt.show()
```

```python
class RidgePenalty:
    
    def __init__(self, l):
        self.l = l
        
    def __call__(self, theta): #__call__ allows us to call class as method
        return self.l * np.sum(np.square(theta))
        
    def derivation(self, theta):
        return self.l * 2 * theta

class Ridge(LogisticRegression):
    def __init__(self, l=0.1, **kwargs):
        kwargs.pop('regularization', None)
        regularization = RidgePenalty(l)
        super().__init__(regularization=regularization, **kwargs)


# class Ridge(LogisticRegression):
#     def __init__(self, method, lr, l=0.1, regularization = Ridgepenalty(l),init_method='xavier', momentum=0.0):
#         self.regularization = RidgePenalty(l)
#         super().__init__(self.regularization, lr, method, init_method=init_method, momentum=momentum)
```
